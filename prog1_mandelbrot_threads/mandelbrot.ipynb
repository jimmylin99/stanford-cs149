```python
import matplotlib.pyplot as plt
```

```python
perf = [
    (1, 1.00),
    (2, 1.99),
    (3, 1.64),
    (4, 2.44),
    (5, 2.50),
    (6, 3.25),
    (7, 3.40),
    (8, 4.02),
]
x = [e[0] for e in perf]
y = [e[1] for e in perf]
```

```python
plt.plot(x, y)
plt.xlabel('thread num')
plt.ylabel('speed up')
plt.show()
```
