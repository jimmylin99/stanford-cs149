```python
with open('result.txt', 'r') as f:
    data = f.readlines()
perf = []
for line in data:
    t, speedup = line.split()
    perf.append((int(t), float(speedup)))
x = [e[0] for e in perf]
y = [e[1] for e in perf]
```

```python
%matplotlib inline
plt.plot(x, y)
plt.xlabel('thread num')
plt.ylabel('speed up')

import os

# Set the filename
filename = 'plot.png'

# Check if the file already exists
if os.path.exists(filename):
    # If it does, append a suffix to the filename
    i = 1
    while os.path.exists(f"{filename[:-4]}_{i}.png"):
        i += 1
    filename = f"{filename[:-4]}_{i}.png"

# Save the plot as a file
# Note: for jupyter notebook (at least DataSpell IDE), there is no need to
# use plt.show and plt.savefig simultaneously, just savefig is fine
plt.savefig(filename)
```
